# Generalized zero point for standard stars

This notebook shows how to use the throughputs for each of the SCORPIO filters to estimate the spectra for a set of standard stars.

To approximate SCORPIO instrumental response curves, I am following the methods laid out by <a href="https://ui.adsabs.harvard.edu/abs/2014A%26A...568A...9M/abstract">Moehler et al. (2014)</a> (more context in cells below).



# Note regarding synphot compatible FITS files for standard star spectra.


This notebook uses XSHOOTER spectra, whose BinTables are formatted in such a way that is incompatible with synphot. I deconstructed the data and created new BinTable files.  The original file names start with `[targname]_[vis/nir/mod]_ADP[...].fits`, new file names are formatted as `[targname]_[vis/nir/mod]_spec.fits`.

The cell below gives an example of how to create the new fits files.  The procedure is simple: extract the arrays for wavelength, flux, and error (if applicable) and enter the data (and an optional new output file name) into the function from `synphote_routines.py`named `create_new_fits_bintab`.



### Make sure the flux and wavelength units are in FLAM and <span>&#8491;</span>!!  

The model XSHOOTER spectra have flux units of 10$\mathrm{mW \ m}^{-2} \mathrm{nm}^{-1}$, so I converted the data to synphot flux units of FLAM $\equiv\mathrm{erg \ s}^{-1} \mathrm{cm}^{-2}$ <span>&#8491;</span>$^{-1}$ before creating the new files.

Flux units for the raw observed data are already in synphot-compatible, so no change there.

Wavelength unit is in nm, so I converted those numbers to <span>&#8491;</span>.




```python
raw_GD152 = "XSHOOTERStandards/GD153_nir_ADP.2017-06-02T18_19_14.331.fits"

phdu_GD152 = fits.open(raw_GD152)[0] # primary extension to be copied to new FITS file.
hdu_GD152 = fits.open(raw_GD152)[1] # the data is in the first extension.


print('wavelength units: ',hdu_GD152.header['TUNIT1']) 
# notice how the wavelength data is in nm, we need to convert this to A.


print('flux units: ',hdu_GD152.header['TUNIT2']) # show flux unit

print('weirdly-shaped data array: ',hdu_GD152.data['WAVE']) 
# also notice how the data array has a weird shape of an array inside an array, 
# which is why I extract the data like data['WAVE'][0] instead of just data['WAVE']. 

wl_ang = (hdu_GD152.data['WAVE'][0]*u.nm).to(u.angstrom) # making sure to convert to angstrom!!!
flx_dat = (hdu_GD152.data['FLUX'][0]*units.FLAM)
flx_err = (hdu_GD152.data['ERR'][0]*units.FLAM)

# once you have the new data arrays, use the 'create_new_fits_bintab' function in synphot_routines.py.
# you can include a new file name to be saved, or just create it for your local notebook.
# since I have already created the new spec files, I am leaving the 'new_fname' argument as None.
new_GD152_bintab = SYNR.create_new_fits_bintab(lambda_angstrom=wl_ang,flx_data=flx_dat,err_data=flx_err,
                                          primary_hdu=phdu_GD152,new_fname=None)
```

# Vega 

First, estimate the filter spectra for Vega.

The spectrum is loaded via `vega = SourceSpectrum.from_vega()`


```python
#read STSDAS vega spectrum

vega = SourceSpectrum.from_vega()
# this is a "SourceSpectrum", synphot object, with units
# in photlam, the default for SourceSpectrum

print("from the original file")
print(max(vega(vega.waveset)))
#Plot using the SourceSpectrum.plot method, specifying different units
vega.plot(left=1000,right=50000,flux_unit='fnu')
vega.plot(left=1000,right=50000,flux_unit='photlam')
vega.plot(left=1000,right=50000,flux_unit='flam')
vega.plot(left=1000,right=50000,flux_unit='Jy')
```

```python
#How to convert units
#First, convert the flux *arrays* from photlam to fnu, flam, fJy
vega_fnu=units.convert_flux(vega.waveset,fluxes=vega(vega.waveset),out_flux_unit = units.FNU)
vega_flam=units.convert_flux(vega.waveset,fluxes=vega(vega.waveset),out_flux_unit = units.FLAM)
vega_Jy=units.convert_flux(vega.waveset,fluxes=vega(vega.waveset),out_flux_unit = u.Jy)
                           
#check:
print("\n from converted fluxes")
print(max(vega_fnu))
print(max(vega_flam))
print(max(vega_Jy))
#ok
                           
#then rebuild the spectra with these these values
vega_fnu=SourceSpectrum(Empirical1D,points=vega.waveset,lookup_table=vega_fnu)
vega_flam=SourceSpectrum(Empirical1D,points=vega.waveset,lookup_table=vega_flam)
vega_Jy=SourceSpectrum(Empirical1D,points=vega.waveset,lookup_table=vega_Jy)
                       
# finally create synphot spectra that have the right flux units.
vega_fnu=vega_fnu(vega.waveset,flux_unit=units.FNU)
vega_flam=vega_flam(vega.waveset,flux_unit=units.FLAM)
vega_Jy=vega_Jy(vega.waveset,flux_unit=u.Jy)
#check:
print("\n from the rebuilt spectra")
print(max(vega_fnu))
print(max(vega_flam))
print(max(vega_Jy))
```

# The get_synphot_table routine

This routine builds spectra for each of the filters and returns a table with the following columns:

<b>Filter name</b> : names with "_no_adc" refer to throughputs with no atmosheric correction <br>
<b>Counts/s</b> :  predicted count rate for the telescope area through the given filter<br>
<b>VegaMag</b> :   predicted Vega magnitude through the given filter (Vega spectrum is in ABmag units<br>
<b>pivot_w</b>l :  pivot wavelength to allow exact conversion between flux densities $F_{\nu}(R)$ and $F_{\lambda}(R)$ (source-independent)<br>
<b>effective_wl</b> :  effective wavelength or the mean wavelength of detected photons (source-dependent)


### $\mathrm{vegamag} = -2.5 \mathrm{log_{10}} \frac{f}{f(VEGA)}$ 

### $\mathrm{ABmag} = -2.5 \mathrm{log_{10}} \frac{f_{\nu}}{3631 \ \mathrm{Jy}} $, unit is FNU



The second item returned is a dictionary where each filter containing the newly built spectra of the standard star.


# XSHOOTER Stars


Observed spectra were retrieved from the <a href="http://archive.eso.org/wdb/wdb/adp/phase3_spectral/form?"> ESO Data Archive</a>.  <a href="http://www.eso.org/rm/api/v1/public/releaseDescriptions/70">Here is the XSHOOTER spectral library data description.</a>

Model spectra for the standards are from <a href="https://ui.adsabs.harvard.edu/abs/2014A%26A...568A...9M/abstract">Moehler et al. (2014)</a>.  The wavelength coverage for this set of stars ($\lambda \ \simeq$ 4000-25000 <span>&#8491;</span>) is similar to SCORPIO's coverage. 


Moehler+2014 describes their procedure to get the XSHOOTER instrumental response as follows:
    Compute the ratio of the observed spectra to the model spectra.  However, because of small deficiencies in the model spectra, most of these ratios show star-specific features in regions with overlapping lines.  To identify and fit these discrepancies, they used a star with no line features in the spectral range, which could also be accurately modelled.  Use the observed/model ratio of that star, along with the observations of the other stars to derive corrections to the other model spectra.
    
Though we do not yet have SCORPIO observations yet, we are trying to replicate their work so that it can be applied to SCORPIO later.  Also, we can at least use the model of these stars to derive the shape of the response curve for now.
